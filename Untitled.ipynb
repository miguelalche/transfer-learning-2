```python
# -*- coding: utf-8 -*-
import pandas as pd
import sys
sys.path.append('/usr/local/lib/python2.7/dist-packages')
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import random
```

```python
def plotMode(mode):

    val_loss = pd.read_csv(mode + '/val_loss.txt', delimiter=";")
    val_accuracy = pd.read_csv(mode + '/val_accuracy.txt', delimiter=";")
    train_loss = pd.read_csv(mode + '/train_loss.txt', delimiter=";")
    train_accuracy = pd.read_csv(mode + '/train_accuracy.txt', delimiter=";")


    accuracy = pd.DataFrame({'val_accuracy': val_accuracy['Accuracy'], 
                    'train_accuracy': train_accuracy['Accuracy']})


    loss = pd.DataFrame({'val_loss':val_loss['Loss'], 'train_loss': train_loss['Loss']})
    
    ax = accuracy.plot(logy=False, figsize=[12,12])
    plt.title("Accuracy for {}".format(mode))
    plt.show()

    ax = loss.plot(logy=False, figsize=[12,12])
    plt.title("Loss for {}".format(mode))
    plt.show()
```

```python
plotMode('fresh')
```

```python
df = pd.DataFrame({})
for mode in ['fresh','feature_extractor', 'finetuning']:
    val_loss = pd.read_csv(mode + '/val_loss.txt', delimiter=";")
    val_accuracy = pd.read_csv(mode + '/val_accuracy.txt', delimiter=";")
    train_loss = pd.read_csv(mode + '/train_loss.txt', delimiter=";")
    train_accuracy = pd.read_csv(mode + '/train_accuracy.txt', delimiter=";")
    
#     df[mode + '_' + 'train_loss'] = train_loss['Loss']
    df[mode + '_' + 'val_loss'] = val_loss['Loss']

#     df[mode + '_' + 'train_accuracy'] = train_accuracy['Accuracy']
    df[mode + '_' + 'val_accuracy'] = val_accuracy['Accuracy']
    
    
ax = df.plot(logy=False, figsize=[12,12])
plt.title('Comparaciones')
plt.show()
```
